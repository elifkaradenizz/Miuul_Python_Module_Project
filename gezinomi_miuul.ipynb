

*   SaleCheckInDayDiff değişkenini kategorik bir değişkene çevirme.



```python
bins = [-1, 7, 30, 90, df[ 'SaleCheckInDayDiff'].max()]
labels= ["Last Minuters", "Potential Planners", "Planners", "Early Bookers"]
```

```python
df["EB_Score"]= pd.cut(df["SaleCheckInDayDiff"], bins, labels = labels)
df.head(50).to_excel ("eb_scorew.xlsx", index=False)
```



*  City-Concept-EB Score, City-Concept-Season, City-Concept-CInDay kırılımında ortalama ödenen ücret ve yapılan işlem sayısı cinsinden
inceleme.



```python
#Sehir-Concept-EB Score kırılımında ücret ortalamaları
df.groupby(by=["SaleCityName", "ConceptName", "EB_Score"]).agg({"Price": ["mean", "count"]})
#Sehir-Concept-Sezon kırılımında ücret ortalamaları
df.groupby(by=["SaleCityName", "ConceptName", "Seasons"]).agg({"Price": ["mean", "count"]})
#Sehir-Concept-CInDay kırılımında ücret ortalamaları
df.groupby(by=["SaleCityName", "ConceptName", "CInDay"]).agg({"Price": ["mean", "count"]})
```



*   City-Concept-Season kırılımının çıktısını PRICE'a göre sıralama.


```python
agg_df = df.groupby(by=["SaleCityName", "ConceptName", "Seasons"]).agg({"Price": "mean"}).sort_values("Price", ascending=False)
agg_df.head(50)
```



*   Indekste yer alan isimleri değişken ismine çevirme.



```python
agg_df.reset_index(inplace=True)
agg_df.head(50)
```



*   Yeni seviye tabanlı müşterileri (persona) tanımlama. Örn: DIĞER_HERŞEY DAHIL_HIGH.



```python
agg_df['sales_level_based'] = agg_df[["SaleCityName", "ConceptName", "Seasons"]].agg(lambda x: '_'.join(x).upper(), axis=1)

# İlk 50 satırı gösterme
agg_df.head(50)
```



*  Yeni müşterileri (personaları) segmentlere ayırma.



```python
agg_df['SEGMENT']= pd.qcut(agg_df['Price'], 4, labels= ['D', 'C', 'B', 'A'])
agg_df.head(20)
```



*  Segmentleri betimleme (Segmentlere göre group by yapıp price mean, max, sum’larını alma).



```python
agg_df.groupby("SEGMENT").agg({"Price": ["mean", "max", "sum"]})
```



*   Antalya’da herşey dahil ve yüksek sezonda tatil yapmak isteyen bir kişinin ortalama ne kadar gelir kazandırması beklenir?



```python
new_user = "ANTALYA_HERŞEY DAHIL_HIGH"
agg_df [agg_df['sales_level_based']== new_user]
```



*   Girne’de yarım pansiyon bir otele düşük sezonda giden bir tatilci hangi segmentte yer alacaktır?



```python
new_user = "GIRNE_YARIM PANSIYON_LOW"
agg_df [agg_df['sales_level_based']== new_user]
```
