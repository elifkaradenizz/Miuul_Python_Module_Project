

*   Çıktıyı PRICE’a göre sıralama.



```python
agg_df= df.groupby(['COUNTRY', 'SOURCE', 'SEX', 'AGE']).agg({'PRICE': 'mean'}).sort_values('PRICE', ascending=False)
agg_df.head(50)
```



*  Indekste yer alan isimleri değişken ismine çevirme.




```python
agg_df.reset_index(inplace=True)
agg_df.head(50)
```



*  Age değişkenini kategorik değişkene çevirip agg_df’e ekleme.



```python
bins = [0, 18, 23, 30, 40, 70]
labels = ['0_18', '19_23', '24_30', '31_40', '41_70']

# Age sütununu kategorik hale getirme ve Age_Cat sütununu oluşturma
agg_df['AGE_CAT'] = pd.cut(agg_df['AGE'], bins, labels=labels)

# İlk 50 satırı gösterme
agg_df.head()
```



*   Yeni seviye tabanlı müşterileri (persona) tanımlama.



```python
agg_df['customers_level_based'] = agg_df[["COUNTRY", "SOURCE", "SEX", "AGE_CAT"]].agg(lambda x: '_'.join(x).upper(), axis=1)

# İlk 50 satırı gösterme
agg_df.head(50)
```



*  customers_level_based değerleri oluşturulduktan sonra bu değerlerin tekilleştirilmesi.



```python
agg_df = agg_df.groupby('customers_level_based').agg({'PRICE': 'mean'})
```

```python
agg_df.reset_index(inplace= True)
agg_df.head()
```

```python
agg_df['customers_level_based'].value_counts()
```


*   Yeni müşterileri (personaları) segmentlere ayırma.



```python
agg_df['SEGMENT']= pd.qcut(agg_df['PRICE'], 4, labels= ['D', 'C', 'B', 'A'])
agg_df.head(20)
```
